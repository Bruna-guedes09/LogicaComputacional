## Socorram-me, subi no ônibus em Marrocos!



**Problema**: checar se uma string é um palíndromo.

**Entrada**: uma string com pelo menos um caractere representando apenas uma palavra.

**Saída**: `True` se a string de entrada for um palíndromo e `False` caso não for. Desconsiderar diferença entre letras maiúsculas e minúsculas (logo, &ldquo;Ana&rdquo; é um palíndromo).

**Nota**: é possível (porém não é obrigatório) resolver esse problema utilizando recursão. A sugestão é tentar resolver por recursão para treinar. Não é necessário entregar duas resoluções (uma com e uma sem recursão). No caso de duas funções com o mesmo nome, apenas a última que for definida será considerada para fins de pontuação.

**Alguns exemplos de entradas e retornos**:


| Entrada <code>palavra</code>|Retorno|
|---|---|
| &ldquo;ovo&rdquo;|True|
| &ldquo;pato&rdquo;|False|
| &ldquo;arara&rdquo;|True|
| &ldquo;Ana&rdquo;|True|
| &ldquo;Natan&rdquo;|True|
| &ldquo;torresmo&rdquo;|False|
| &ldquo;sopapos&rdquo;|True|

**Algoritmo**:



```python
def checa_palindromo(palavra):

    palavra_padrao = palavra.lower().replace(" ", "") #remove espaços entre as palavras e deixa as letras minúsculas
    
    #caso base
    if len(palavra_padrao) <= 1:
        return True
    #parte recursiva     
    elif palavra_padrao[0] == palavra_padrao[-1]:
        nova_palavra = palavra_padrao[1:-1]
        return checa_palindromo(nova_palavra) 
    else:
        return False
        
#lista para testar os valores
lista_palavras = ["Socorram me subi no onibus em Marrocos", "ovo", "Ovo", "Pato", "patO", "pato", "arara", "Ana", "Natan", "torresmo", "Torresmo", "sopapos"] 

#loop para enviar todos os valores da lista para a função, exibindo os valores de entrada e saída da função
for palavra in lista_palavras: 
    print("Entrada: ", palavra)
    print("Saída: ", checa_palindromo(palavra)) #enviamos o grafo para a função
    print()
    print("---------------------------------------- ") #linha divisória, por fins de organização
```

## O máximo divisor comum



**Problema**: encontrar o máximo divisor comum entre dois números inteiros $a$ e $b$.

**Entrada**: dois números inteiros $a$ e $b$, tal que $a \geq 0$ e $b \geq 0$.

**Saída**: número inteiro representando o máximo divisor comum entre os dois números de entrada, $\mathrm{MDC}(a,b)$.

**Definição 1**: se $b=0$, então $\mathrm{MDC}(a,b) = a$.

**Definição 2**: se $b \neq 0$, então $\mathrm{MDC}(a,b) = \mathrm{MDC}(b,c)$, onde $c$ é o resto da divisão de $a$ por $b$.

**Nota**: é possível (porém não é obrigatório) resolver esse problema utilizando recursão. A sugestão é tentar resolver por recursão para treinar. Não é necessário entregar duas resoluções (uma com e uma sem recursão). No caso de duas funções com o mesmo nome, apenas a última que for definida será considerada para fins de pontuação.

**Alguns exemplos de entradas e retornos**:


| Entrada <code>a</code>|Entrada <code>b</code>|Retorno|
|---|---|---|
| 0|0|0|
| 1|0|1|
| 0|10|10|
| 10|5|5|
| 50|75|25|
| 48|18|6|
| 10|2|2|
| 101|103|1|



```python
def mdc(a, b):

    if a == 0 and b == 0:
        return 0
    elif b == 0:
        return abs(a)  #para que resultado seja sempre positivo
    else:
        c = a % b
        return mdc(b, c)

#lista para testar os valores
lista_a = [0, 1, 0, 10, 50, 48, 10, 101, -6, 9, 7]
lista_b = [0, 0, 10, 5, 75, 18, 2, 103, -10, -6, -77] 

#loop para enviar todos os valores da lista para a função, exibindo os valores de entrada e saída da função
for a, b in zip(lista_a, lista_b): 
    print("Entrada a: ", a, "Entrada b: ", b)
    print("Saída: ", mdc(a, b)) #enviamos os números para a função
    print()
    print("---------------------------------------- ") #linha divisória, por fins de organização
```

## Soma dígitos



**Problema**: dado um número inteiro $n \geq 0$, encontrar a soma de seus dígitos. Exemplo: 85 resulta em 8+5=13.

**Entrada**: número inteiro $n \geq 0$.

**Saída**: número inteiro representando a soma dos dígitos de $n$.

**Dica**: vale a pena revisar como funcionam os operadores `//` e `%` de Python.

**Nota**: é possível (porém não é obrigatório) resolver esse problema utilizando recursão. A sugestão é tentar resolver por recursão para treinar. Não é necessário entregar duas resoluções (uma com e uma sem recursão). No caso de duas funções com o mesmo nome, apenas a última que for definida será considerada para fins de pontuação.

**Alguns exemplos de entradas e retornos**:


| Entrada <code>n</code>|Retorno|
|---|---|
| 0|0|
| 1|1|
| 10|1|
| 85|13|
| 100|1|
| 150|6|
| 9999|36|
| 1234567890|45|

**Algoritmo**:



```python
def soma_digitos(n):

    #caso base
    if (n/10) < 1:
        return n
    else:
        #retorna o módulo divisão por 10 + o resto do número. n//10 remove o último dígito, n%10 pega o último dígito
        return n % 10 +  soma_digitos(n//10)
        
#lista para testar os valores
lista = [0, 1, 10, 85, 100, 150, 9999, 1234567890]

#loop para enviar todos os valores da lista para a função, exibindo os valores de entrada e saída da função
for n in lista:
    print("Entrada: ", n)
    print("Saída (com recursão): ", soma_digitos(n)) #enviamos o número para a função
    print()
    print("---------------------------------------- ")  #linha divisória, por fins de organização
```

## A sequência de Recamán



**Problema**: gerar a sequência de Recamán até o termo  $a_n$.

**Entrada**: número inteiro $n \geq 0$.

**Saída**: lista iniciando com o valor $a_0$ e indo até o valor $a_n$ seguindo a sequência de Recamán.

**Material para leitura**: a sequência de Recamán $a_0$, $a_1$, $a_2\ldots$ começa a contar para $n=0$ e cada termo $a_n$ é definido como:

$a_{n}\begin{cases}
0 & \textrm{se }n=0\\
a_{n-1}-n & \textrm{se }a_{n-1}-n>0\textrm{ e se }a_{n-1}-n\textrm{ ainda não está na sequência}\\
a_{n-1}+n & \textrm{caso as outras condições não sejam satisfeitas}
\end{cases}$

Os primeiros números da sequência são: 0, 1, 3, 6, 2, 7, 13, 20, 12, 21, 11, 22, 10, 23, 9, 24, 8, 25, 43, 62, 42, 63, 41, 18, 42, 17, 43, 16, 44, 15, 45, 14, 46, 79, 113, &#x2026;

**Nota**: é possível (porém não é obrigatório) resolver esse problema utilizando recursão. A sugestão é tentar resolver por recursão para treinar. Não é necessário entregar duas resoluções (uma com e uma sem recursão). No caso de duas funções com o mesmo nome, apenas a última que for definida será considerada para fins de pontuação.

**Alguns exemplos de entradas e retornos**:


| Entrada <code>n</code>|Retorno|
|---|---|
| 0|[0]|
| 1|[0, 1]|
| 2|[0, 1, 3]|
| 5|[0, 1, 3, 6, 2, 7]|
| 10|[0, 1, 3, 6, 2, 7, 13, 20, 12, 21, 11]|

**Algoritmo**:



```python
#nessa função, precisei da ajuda da IA, para entender quais eram os meus erros na hora de montar a estrutura recursiva
#Nesse contexto, a ideia de criar uma outra função interna recursiva foi da IA. A explicação foi: "Mas se você só tem n como argumento, não há onde guardar ou passar esses valores entre as chamadas recursivas, já que cada chamada precisa conhecer o histórico da sequência."

def recaman(n): 

    #função interna auxiliar que constrói a sequência
    def rec(k, seq_recaman):

        #caso base. Dessa forma, k vai até n
        if len(seq_recaman) > n:
            return seq_recaman

        #se a subtração seq_recaman[-1] - k é positiva e o elemento não está na lista
        elif seq_recaman[-1] - k > 0 and (seq_recaman[-1] - k) not in seq_recaman:
            proximo = seq_recaman[-1] - k
            
        #Caso contrário, soma k ao último elemento
        else: 
            proximo = seq_recaman[-1] + k

        #Adiciona o próximo termo à sequência
        seq_recaman.append(proximo)
        
        #chamada recursiva com k+1 e a nova sequência
        return rec(k + 1, seq_recaman)
        
    #Chamada inicial da função recursiva, começando com k=1 e sequência iniciada em [0]
    return rec(1, [0])


lista = [0, 1, 2, 5, 10, 15, 44]

for n in lista:
    print("Entrada: ", n)
    print("Saída (com recursão): ", recaman(n))
    print()
    print("---------------------------------------- ")
```
